```python
def extract_age_range(age_group):
    if age_group == '70 or Older':
        return 70, None  # Min age is 70, no max age
    elif age_group == '50 to 69':
        return 50, 69
    elif age_group == '30 to 49':
        return 30, 49
    elif age_group == '18 to 29':
        return 18, 29
    elif age_group == '0 to 17':
        return 0, 17
    else:
        return None, None  # Handle any unexpected cases

# Apply the function to the Age Group column and create new columns for min and max age
df[['Min Age', 'Max Age']] = df['age_group'].apply(lambda x: pd.Series(extract_age_range(x)))

print(df)
```
